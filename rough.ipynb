```python
import pandas as  pd
```

```python
df=pd.read_csv("ds.csv",index_col=0)[::-1]
df
```

```python
df.columns
```

```python
df_new=df[['Date',  'Open', 'High', 'Low','Price', 'Vol.', 'Change %']]
```

```python
df_new.reset_index(drop=True)
df_new
```

```python
df_new.rename(columns={"Price": "Close"}, inplace=True)
```

```python
df_new
```

```python
# calculate ATR for df_new
```

```python
df_new['high_low'] = df_new["High"] - df_new["Low"]
df_new["high_close"]= abs(df_new["High"] - df_new["Close"].shift(1))
df_new["low_close"]= abs(df_new["Low"] - df_new["Close"].shift(1))
```

```python
df_new
```

```python
df_new["TR"] = df_new[["high_low", "high_close", "low_close"]].max(axis=1)
df_new
```

```python
ATR=df_new["TR"].mean()
ATR
```

```python
# function to calculate standard deviation of "col" of last n days
def std_dev(df,col,n):
    return df[col].tail(n).std()
```

```python
std_dev(df_new,"Close", 2)
```

```python
df_new["ATR"]=df_new["TR"].rolling(window=len(df_new)).mean()
```

```python
df_new['ATR']=df_new["TR"].expanding().mean()
df_new
```

```python
std_dev(df_new,"Close", 10)
```

```python
std_dev(df_new,"ATR", 20)
```

```python
std_dev(df_new,"TR", 20)
```

```python
df_new.to_csv("ds_new.csv")
```
